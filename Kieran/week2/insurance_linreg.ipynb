```python
insurance = pd.read_csv('insurance.csv')
insurance.head()
```

```python
def convert_to_binary(sm):
    if sm == "yes":
        return 1
    elif sm == "no":
        return 0

insurance["binary_smoker"] = insurance["smoker"].apply(convert_to_binary)
```
